```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
df=pd.read_csv("C:/Users/Hp/Downloads/amazon sale report.csv")
amazon=df
amazon
```

```python
amazon.info()
```

```python
amazon.isnull().sum()
```

```python
amazon.drop(columns=['Unnamed: 22','fulfilled-by','ship-country','currency','Sales Channel '],inplace=True)
amazon
```

```python
amazon.drop_duplicates(subset=['Order ID','ASIN'],inplace=True)
amazon
```

```python
amazon.fillna(value={'Courier Status':'Unknown','promotion-ids':'no promotion','Amount':0,'ship-city':'not available',
              'ship-state':'not available','ship-postal-code':'not available'},inplace=True)
amazon
```

```python
amazon.isnull().sum()
```

```python
amazon.rename(columns={'B2B':'Customer_type','Amount':'Order_amount(₹)'},inplace=True)
amazon
```

```python
amazon['Customer_type'].replace(to_replace=[True,False],value=['Business','Customer'],inplace=True)
amazon
```

```python
amazon['order_amount_($)'] = amazon['Order_amount(₹)'].apply(lambda x: x *0.0120988)
amazon
```

```python
amazon["Date"]=pd.to_datetime(amazon["Date"])
amazon
```

```python
amazon.info()
```

```python
mask = (amazon['Date'] >= '04-01-22')
amazon=amazon.loc[mask]
amazon
```

```python
amazon['Month'] = amazon['Date'].dt.month_name(locale='English')
amazon
```

```python

amazon.replace({'ship-state': {'AR': 'ARUNACHAL PRADESH', 'NL': 'NAGALAND','orissa':'ODISHA','PB':'PUNJAB',
                               'Punjab/Mohali/Zirakpur':'PUNJAB','pondicherry':'PUDUCHERRY',
                               'Rajshthan':'RAJASTHAN','rajsthan':'RAJASTHAN','RJ':'RAJASTHAN',
                               
}},inplace=True)
```

```python
amazon['ship-state'] = amazon['ship-state'].replace(['AR','Arunachal Pradesh','Arunachal pradesh'], 'ARUNACHAL PRADESH')
amazon['ship-state'] = amazon['ship-state'].replace(['NL','Nagaland'], 'NAGALAND')
amazon['ship-state'] = amazon['ship-state'].replace(['orissa','Odisha','Orissa'], 'ODISHA')
amazon['ship-state'] = amazon['ship-state'].replace(['PB','Punjab/Mohali/Zirakpur','punjab','Punjab'], 'PUNJAB')
amazon['ship-state'] = amazon['ship-state'].replace(['pondicherry','Puducherry','Pondicherry'], 'PUDUCHERRY')
amazon['ship-state'] = amazon['ship-state'].replace(['Rajshthan','rajsthan','RJ','Rajasthan','Rajsthan','rajasthan'], 'RAJASTHAN')
amazon['ship-state'] = amazon['ship-state'].replace(['Bihar','bihar'], 'BIHAR')
amazon['ship-state'] = amazon['ship-state'].replace(['Chandigarh'], 'CHANDIGARH')
amazon['ship-state'] = amazon['ship-state'].replace(['Delhi','New Delhi','delhi'], 'DELHI')
amazon['ship-state'] = amazon['ship-state'].replace(['Goa','goa'], 'GOA')
amazon['ship-state'] = amazon['ship-state'].replace(['Manipur'], 'MANIPUR')
amazon['ship-state'] = amazon['ship-state'].replace(['Mizoram'], 'MIZORAM')
amazon['ship-state'] = amazon['ship-state'].replace(['Sikkim'], 'SIKKIM')
amazon['ship-state'] = amazon['ship-state'].replace(['Meghalaya'], 'MEGHALAYA')
```

# PRELIMINARY ANALYSIS

```python
t_r=amazon.groupby('Month')["Order_amount(₹)"].sum()
print(t_r)
print('\n')
a_j=t_r['April']-t_r['June']
t_r_p=(a_j/t_r['April'])*100
print(f' decreased revenue from april to june :{t_r_p:.2f}')
print('\n')
d_r_a_m=t_r['April']-t_r['May']
d_r_a_m_p=(d_r_a_m/t_r['April'])*100
print(f'decreased revenue from april to may :{d_r_a_m_p:.2f}')
print('\n')
d_r_m_j=t_r['May']-t_r['June']
d_r_m_j_p=(d_r_m_j/t_r['May'])*100
print(f' decreased revenue from may to june : {d_r_m_j_p:.2f}')
```

```python
r_s=amazon.groupby('Category')['Order_amount(₹)'].sum()
a=r_s.sum()
s_r_a=r_s['Set']/a*100
print(f'total set revenue :{s_r_a:.2f}')
k_r_a=r_s['kurta']/a*100
print('\n')
print(f'total kurta revenue :{k_r_a:.2f}')
print('\n')
wd_r_a=r_s['Western Dress']/a*100
print(f'total western dress revenue :{wd_r_a:.2f}')
```

```python
t_r=amazon.groupby(['Category'],as_index=False)['order_amount_($)'].mean().sort_values(by='order_amount_($)',ascending=False)
t_v = t_r.sort_values(by='order_amount_($)', ascending=False).head()
t_v
```
